## 데이터 정리
- 제안자, 의안번호, 대수, 제안자목록링크 컬럼 삭제
- 'null'문자 null값으로 변경
- 공동발의자 리스트로 변경
- 법률안명 사이 ':' 대입
- 제안일 타입 datetime으로 변경
- 컬럼 순서 정리

### 1) 컬럼 삭제 및 null값 처리

```python
title_df1 = title_df1.drop(['의안번호', '대수', '제안자', '제안자목록링크'], axis = 1)
title_df1
```

```python
title_df1.info()
```

```python
# null문자 처리
title_df1.replace('null', np.nan, inplace = True)

# 확인 
print(title_df1.isnull().sum())

# 공동발의자 null 값 확인
title_df1[title_df1['공동발의자'].isnull()] #나중에 처리
```

```python
# 공동발의자 null 값 확인
title_df1[title_df1['공동발의자'].isnull()] #나중에 처리
```

```python
# 소관위원회 null값 확인
title_df1[title_df1['소관위원회'].isnull()]

# 총 약 1천 7백개중 null값 4개 (처리상태 철회가 3건이라 drop)
title_df_drop = title_df1.dropna(subset = ['소관위원회'])
```

```python
title_df_drop
```

```python
title_df_drop.isnull().sum()
```

### 2) 공동발의자 리스트 변경

```python
ti_df = title_df_drop
ti_df['공동발의자'] = ti_df['공동발의자'].str.split(',').tolist() #나중에 전처리를 위해 

print(ti_df.info())

#공동발의자 값 확인
ti_df['공동발의자'].value_counts()
```

```python
# 공동발의자 종류 확인
ti_df2 = ti_df.copy()
ti_df2.dropna(subset=['공동발의자'], inplace = True)
li = []
co_li = list(ti_df2['공동발의자'])

li = []
for i in range((len(co_li))):
    for t in range(len(co_li[i])):
        if co_li[i][t] in li:
            pass
        else:
            li.append(co_li[i][t])
print(li, '\n')
print('공동발의자 개수: ', len(li), '개') 
```

```python
# 대표발의자랑 공동발의자 중복확인
# 대표발의자 : per_li, 공동발의자 : li

both_li = []
for i in per_li:
    for t in li:
        if i == t:
            both_li.append(i)
print(both_li, '\n', len(both_li), '명')
```

### 3) 법률안명 처리

```python
# 법률안 종류 확인
ti_df['법률안명'].unique()
```

```python
# 일부개정법률안, 특별법안, 전부개정법률안, 법률안, 일부개정을 위한 법률안 앞에 | 생성 (구분을 위해!)

ti_df1 = ti_df.replace('\s(\w+$)', r' | \1', regex=True ) # \1번은 그대로
ti_df1 = ti_df1.replace('\s(일부개정을\s위한\s)\|\s(법률안$)', r'| \1\2', regex=True ) # \'일부개정을 위한 법률안 수정
ti_df1
```

```python
ti_df1['법률안명'].unique()
```

### 4) 제안일 datetime 변경

```python
ti_df1['제안일'] = pd.to_datetime(ti_df1['제안일'])
ti_df1.info()
```

### 5) 컬럼순서 변경

```python
# 행번호 다시 생성 후 index 삭제
final_df = ti_df1.reset_index()
final_df1 = final_df.drop('index', axis= 1)
final_df1
```

```python
#컬럼순서 정리
final_df1 = final_df1[['법률안명', '대표발의자', '공동발의자', '소관위원회', '제안일', '처리상태', '상세페이지']]
final_df1.info()
```
